#4. Opérations sur le fichier CSV burkina_location.csv

```python
#4.1
#Extraire les données contenant le nom 'gounghin',
df1=pd.read_csv('burkina_location.csv')
gounghin=df1.loc[df1['location_name'].str.contains('Gounghin')]
gounghin.drop(['Unnamed: 0'], axis = 1, inplace=True)
gounghin
```

```python
# enregistrez-le sous le fichier gounghin.csv
gounghin.to_csv('gounghin.csv', index=False, encoding='utf-8')
```

```python
#4.2
#Extraire la sous-partie de la base de données (fichier burkina_location.csv), dont
#les premières lettres des noms de lieux sont compris entre 'A' et 'P' (ordre
#alphabétique).
```

```python
A_P = df1[df1['location_name'].between('A','Q')]
A_P.drop(['Unnamed: 0'], axis = 1, inplace=True)
A_P.sort_values(by=['location_name'], inplace=True)
A_P
```

```python
#4.3.a
#Identifiez respectivement, la latitude,  
#correspondants dans le fichier
min_lat=df0[df0['lat'] == df0['lat'].min()]

min_lat
```

```python
#4.3.b
#la longitude minimale
min_long=df0[df0['long'] == df0['long'].min()]
min_long
```

```python
#4.3.c
#En bonus
#les noms de lieux
min_name=df0[df0['location_name'] == df0['location_name'].min()]
min_name
```

```python
#4.4
#coordonnées sont comprises entre (lat >= 11 et lon <= 0.5)
lat11_long05=df0.loc[(df0['lat'] >= 11) & (df0['long'] <= 0.5)]
lat11_long05
```
